# removing outliers from price

```python
column_name = numerical['price']
price_df=pd.DataFrame(column_name)
price_df
```

```python
sns.boxplot(data=numerical.price)
plt.title('values for the price before removing outliers')
plt.ylabel('points')
plt.xlabel('Price')
plt.show()
```

-boxplot to represent outleirs points of the price .

```python
Q1 =np.percentile(price_df,25)
Q3 =np.percentile(price_df,75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_filtered = price_df[(price_df >= lower_bound) & (price_df <= upper_bound)]
print(Q1 , Q3, IQR ,lower_bound,upper_bound)
```

```python
numerical.price[numerical.price>Q3]=upper_bound
numerical.price[numerical.price<Q1]=lower_bound
sns.boxplot(data=numerical.price)
plt.title('values for the price after removing outliers')
plt.ylabel('points')
plt.xlabel('Price')
plt.show()
```

-boxplot to represent the price after removing outliers .
